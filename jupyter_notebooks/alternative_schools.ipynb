```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_files/frpm.csv')
```

```python
df.head(n=5)
```

```python
df = df[df['high_grade']=='12']
```

```python
print df['educational_option'].unique()
```

```python
choice = 'free_and_reduce_count'
print df['enrollment'].sum()
traditional_count = df[df['educational_option']=='Traditional'][choice].sum()
juvenile_count = df[df['educational_option']=='Juvenile Court School'][choice].sum()
county_count = df[df['educational_option']=='County Community School'][choice].sum()
state_count = df[df['educational_option']=='State Special School'][choice].sum()
alternative_count = df[df['educational_option']=='Alternative School of Choice'][choice].sum()
continuation_count = df[df['educational_option']=='Continuation School'][choice].sum()
special_ed_count = df[df['educational_option']=='Special Education School'][choice].sum()
community_count = df[df['educational_option']=='Community Day School'][choice].sum()
home_hospital_count = df[df['educational_option']=='Home and Hospital'][choice].sum()
opportunity_count = df[df['educational_option']=='Opportunity School'][choice].sum()
youth_authority_count = df[df['educational_option']=='Youth Authority School'][choice].sum()

print traditional_count
print juvenile_count
print alternative_count
print county_count
```

```python
labels = 'Traditional', 'Juvenile Court School', 'County Community School','State Special School', 'Alternative School of Choice', 'Continuation School', 'Special Education School', 'Community Day School', 'Home and Hospital', 'Opportunity School', 'Youth Authority School'
print len(labels)
sizes = [traditional_count, juvenile_count, county_count, state_count, alternative_count, continuation_count, special_ed_count,
         community_count, home_hospital_count, opportunity_count, youth_authority_count]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
